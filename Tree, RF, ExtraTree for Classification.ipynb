```python
cat_to_dummies=['workclass','marital.status', 'occupation','relationship', 'race', 'sex', 'native.country']
simple_numeric=['age', 'fnlwgt', 'education.num', 'capital.gain', 'capital.loss','hours.per.week']
```

```python
dp=dt.DataPipe(cat_to_dummies=cat_to_dummies,
              simple_numeric=simple_numeric)
```

```python
dp.fit(data)
```

```python
x_train=dp.transform(data)
```

```python
data['Y'].value_counts()
```

```python
y_train=(data['Y']==' >50K').astype(int)
```

```python
y_train.value_counts()
```

```python
data['Y'].unique()
```

```python
x_train.head()
```

```python
params_dt={
    'max_depth':[None,5,10,15,20,30,50,70],
    'min_samples_leaf':[1,2,5,10,15,20],
    'min_samples_split':[2,5,10,15,20],
    'class_weight':[None,'balanced'],
    'criterion':['entropy','gini']
}
```

```python
clf_dt=DecisionTreeClassifier()
```

```python
rs_dt=RandomizedSearchCV(clf_dt,
                        cv=10,
                        param_distributions=params_dt,
                        scoring='roc_auc',
                        n_iter=10,
                        n_jobs=-1,
                        verbose=20)
```

```python
rs_dt.fit(x_train,y_train)
```

```python
dt.report(rs_dt.cv_results_,5)
```

```python
display(graphviz.Source(export_graphviz(rs_dt.best_estimator_,feature_names=x_train.columns,filled=True,
                                        rounded=True)).view())
```
