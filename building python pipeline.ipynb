```python
def dir_percent(dir_percent):
    
    dir_num=dir_percent.str.replace('%',"")
    dir_num=pd.to_numeric(dir_num,errors='coerce')
    return dir_num
```

```python
def fico(fico):
    
    k=fico.str.split('-',expand=True).astype(int)
    
    fico_num=0.5*(k[0]+k[1])
    
    return fico_num
    
```

```python
def el(el):
    
    el=el.str.replace('< 1 year','0')
    el=el.str.replace('10+ years','10',regex=False)
    el=el.str.replace('years','',regex=False)
    el=el.str.replace('year','',regex=False)
    
    el_num=pd.to_numeric(el,errors='coerce')
    
    return el_num
```

```python
# 'ID' : ignore 
# 'Amount.Funded.By.Investors' : ignore 
# 'Interest.Rate' : ignore

#'Amount.Requested' : convert to numeric , missing values
# 'Open.CREDIT.Lines' : convert to numeric , missing values
# 'Revolving.CREDIT.Balance' : convert to numeric , missing values


# 'Loan.Length' : missing values , create dummies with freq cutoff 1% or 250 of the data
# 'Loan.Purpose' : missing values , create dummies with freq cutoff 1% or 250 of the data
#'State' : missing values , create dummies with freq cutoff 1% or 250 of the data
#'Home.Ownership' : missing values , create dummies with freq cutoff 1% or 250 of the data

#'Debt.To.Income.Ratio' : custom_func , missing values
# 'FICO.Range' : custom_func , missing values
#'Employment.Length' : custom_func , missing values

# 'Monthly.Income' : missing values handling 
# 'Inquiries.in.the.Last.6.Months' : missing values handling

cat_to_dummies=['Loan.Length','Loan.Purpose','State','Home.Ownership']
simple_numeric=['Monthly.Income','Inquiries.in.the.Last.6.Months']
custom_var_dict={'Debt.To.Income.Ratio':dir_percent,'FICO.Range':fico,'Employment.Length':el}
cat_to_numeric=['Amount.Requested','Open.CREDIT.Lines','Revolving.CREDIT.Balance']
```
